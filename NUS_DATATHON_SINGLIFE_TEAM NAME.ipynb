### Data preprocessing

```python
import random 
def impute(col):
    #col = col.to_numeric()
    ratioTrue = float(col.sum())/float(len(col))
    def mapNull(x):
        if pd.isnull(x):
            returnNum = 0.0
            if random.random() < ratioTrue:
                returnNum = 1.0
            return returnNum
        else:
            return x
        
    return col.map(mapNull)
```

```python
import datetime
def parseDatetime(col):
    col = pd.to_datetime(col,errors='ignore')
    def tdToPresent(dt):
        if isinstance(pd.to_datetime(dt,errors='ignore'), datetime.datetime):
            return (pd.to_datetime("2024-01-01") -  pd.to_datetime(dt)).days
        else:
            return 0
    return col.apply(tdToPresent)
```

```python
def imputeContinuous(col):
    std = col.std
    mean = col.mean
    def mapNull(x):
        if pd.isnull(x):
           return np.random.normal(mean,std,1)[0]
        else:
           return x
    return col.map(mapNull)
```
